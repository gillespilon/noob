# Ideas

- Rename one column
- Rename multiple columns
- Rename columns with a list
- Rename columns with a list read from a file
- Create a new column from operations on one or more other columns
- Delete a column

```python
# Read a csv file.
import pandas as pd
df = pd.read_csv('data/cloquet_two_weeks_60_min.csv',
                 parse_dates=True,
                 index_col='Time')
```

```python
# Show the column names.
df.columns
```

## Rename one column

```python
df = df.rename(index=str,
               columns={"Water Load (lb/MSF)": "Water Load (lb/msf)"})
```

```python
df.head()
```

## Rename multiple columns

```python
df = df.rename(index=str,
               columns={"Water Load (lb/MSF)": "Water Load (lb/msf)",
                   "Trim Board Density (lb/cft)": "Trim Board Density (lb/cubic ft)"})
```

```python
df.head()
```

## Rename multiple columns using a list

```python
NEW_NAMES = {"Trim Board Weight (lb/sft)": "Trim Board Weight (lb/square ft)"}
df = df.rename(index=str,
               columns=NEW_NAMES)
```

```python
df.head()
```

```python
# Rename all columns using a list.
ALL_NAMES = ['y1', 'y2', 'y3', 'y4', 'y5',
             'y6', 'y7', 'y8', 'y9', 'y10',
             'y11', 'y12', 'y13', 'y14', 'y15',
             'y16', 'y17', 'y18', 'y19', 'y20',
             'y21', 'y22', 'y23', 'y24', 'y25',
             'y26', 'y27', 'y28', 'y29']
df.columns = ALL_NAMES
```

```python
df.head()
```

## Rename columns with a list read from a file

```python
df.columns = pd.read_table('data/more_names.csv',
                           header=None)[0].tolist()
```

```python
df.head()
```

## Create a new column from operations on one or more other columns

```python
df['Wool Difference (%)'] = df['a5'] - df['a6']
```

```python
df.head()
```

## Delete a column

```python
df = df.drop(columns=['Wool Difference (%)'])
```

```python
df.head()
```
